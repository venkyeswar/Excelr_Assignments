```python
# from the above statistical description we can analyse that in few columns oultliers are present.
# drawing the boxplot for every column.
columns = list(df.columns)
for column in columns:
    plt.figure(figsize=(8, 6)) 
    plt.boxplot(df[column]) 
    plt.title(f'Boxplot for column: {column}')
    plt.ylabel('Value')
    plt.show()
    

```

 From the above graphs there are large number of outliers in AC,FM,DL,DS,DP,MSTV,ALTV,MLTV,NSP ,, So treating them may cause error in the model building , so treate the remaining columns and keep these columns without change and build the model if the outliers influncing the model accuracy then treate them.

```python
# treating the outliers in LB column

def HandleOutliers(column,df):
    q1 = df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    lower_boundry=q1-1.5*iqr
    upper_boundry = q3+1.5*iqr
    for value in df[column]:
        if value > upper_boundry:
            df[column].replace(value,upper_boundry,inplace=True)
        elif value<lower_boundry:
            df[column].replace(value,lower_boundry,inplace=True)
        else:
            pass


columns =["LB","UC","ASTV","Width","Tendency"]

for column in columns:
    HandleOutliers(column,df)
```

```python

for column in columns:
    figure=plt.figure(figsize=(8,5))
    plt.boxplot(df[column])
    plt.title(column)
    plt.show()

# hence outlier are Handled for the possible columns.
```
