## Task 1 : Data Preprocessing 

```python
# Import necessary libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
data = pd.read_excel("online retail.xlsx")
```

```python
data
```

```python
data.columns =["items"]
```

```python
data.info()
```

```python
# So the data has no null values.
# we have to use transaction encoder for converting the transaction data into appropriate format.

# Converting the data into list of transactions.
Data = [ trans.split(',') for trans in data['items'] ]
```

```python
Data
```

```python
# Applying Transaction Encoder
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
```

```python
te_array = te.fit(Data).transform(Data)
```

```python
te_array
```

```python
df = pd.DataFrame(te_array,columns = te.columns_)
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

## Task 2 : Association Rule Mining

```python
# import apriori algorithm and find the frequent_itemsets

from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(df,min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=0.5)
```

```python
rules.sort_values(by = "confidence")
```

## Interview questions : 
### 1.	What is lift and why is it important in Association rules?

   **Lift** : lift is the ratio of the observed support to that expected if X and Y are independent .
   it measures how much more likely the consequent is given the antcecedent compared to its typical occurances.

### 2. what is support and Confidence, how do you calculate them ?

   **Support** : measures how frequently the itemset appears in the dataset.

   **support** (X) = no.of transactions contain X/ total no.of transactions.

   **Confidence** : the probability that a transaction containing antecedent also contains the consequent.

   **confidence**(x-->y) = support(XUY)/support(x)

### 3. What are some limitations or challenges of Association mining ?

1. Too Many Rules : It generates large number of rules, many of them are irrelevant and uninteresting.
2. Support for Rare Items : items with low frequency may be overlooked, even if they have strong associations.
3. Difficulty in Setting Threshold values.
