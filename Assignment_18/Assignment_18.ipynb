## Task 2 : Model Implementation

```python
# Split the data into train and test.
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(X,Y_labels,test_size = 0.2 , random_state = 7)
```

```python
print("train_x",train_x.shape)
print("train_y",train_y.shape)
print("test_x",test_x.shape)
print("test_y",test_y.shape)
```

```python
# Model building 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def create_model(optimizer="adam",activation="relu"):
    model = Sequential()
    model.add(Dense(64,input_dim=train_x.shape[1],activation=activation))
    model.add(Dense(32,activation=activation))
    model.add(Dense(Y_labels.shape[1],activation="softmax"))
    model.compile(loss="categorical_crossentropy",optimizer=optimizer ,metrics = ['accuracy'])
    return model
```

```python
model = create_model()
```

```python
model.fit(train_x,train_y,epochs=12,batch_size=30,verbose=1)
```

```python
# Evaluation
y_pred = model.predict(test_x)
```

```python
y_pred
```

```python
y_pred_classes =np.argmax(y_pred,axis=1)
y_pred_classes
```

```python
y_classes = np.argmax(test_y,axis=1)
```

```python
y_classes
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_classes,y_pred_classes))
```

```python
print(accuracy_score(y_classes,y_pred_classes)*100)
```

## Task 3 : Hyperparameter Tuning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier

def build_model(optimizer='adam', **kwargs):
    model = Sequential()
    model.add(Dense(units=64, input_shape=(16,), activation='relu'))
    model.add(Dense(units=32, activation='relu'))
    model.add(Dense(units=26, activation='softmax'))
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python

model = KerasClassifier(model=build_model, epochs=10, batch_size=32)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'model__optimizer': ['adam', 'rmsprop'],
    'epochs': [10, 20],
    'batch_size': [32, 64]
}

grid_model = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)

# Fit the model
grid = grid_model.fit(train_x, train_y)


```

```python
# Retriving the best parameters from the grid.
tuned_model = grid.best_estimator_
print("Best Parameters: ", grid.best_params_)
```

```python
tuned_model
```

```python
y_pred = tuned_model.predict(test_x)
y_pred_classes = np.argmax(y_pred,axis=1)
```

```python
y_pred_classes
```

```python
print(accuracy_score(y_classes,y_pred_classes)*100)
```

```python
print(classification_report(y_classes,y_pred_classes))
```
